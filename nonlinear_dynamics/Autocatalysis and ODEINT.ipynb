```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
k1 = 0.01; k2 = 0.1
Cs = [0.1, 0.1]
```

```python
def get_Cs(C, t):
    A, X = C
    dAdt = -k1*A*X - k2*X**2
    dXdt = k1*A*X - k2*X**2
    return [dAdt, dXdt]

times = np.linspace(0., 100., 101)
ans = odeint(func=get_Cs, y0=Cs, t=times)
As = ans[:, 0]
Xs = ans[:, 1]

fig, ax = plt.subplots()
ax.plot(times, As, label='A')
ax.plot(times, Xs, label='X')
ax.legend()
ax.set_xlabel('time, hrs')
ax.set_ylabel('Concentration, M')
plt.grid()
```

```python
def get_dXdt(X):
    A = 1.
    k1 = 1.; k2 = 1.

    dAdt = -k1*A*X - k2*X**2
    dXdt = k1*A*X - k2*X**2
    return dXdt

Xs = np.linspace(-1, 2., 101)
dXdts = [get_dXdt(X) for X in Xs]

fig, ax = plt.subplots()
ax.plot(Xs, dXdts, label='X')
ax.plot(0,0, 'ro')
ax.plot(1,0, 'go')
ax.legend()
plt.grid()
```

```python
fig, ax = plt.subplots()
ax.set_xlabel('time, hrs')
ax.set_ylabel('Concentration, M')
plt.grid()

k1s = np.linspace(0., 1., 101)
k2s = np.linspace(0., 1., 101)

for const1, const2 in zip(k1s, k2s):
    k1 = const1
    k2 = const2
    times = np.linspace(0., 100., 101)
    ans = odeint(func=get_Cs, y0=Cs, t=times)
    As = ans[:, 0]
    Xs = ans[:, 1]
        
    
    ax.plot(times, As, 'k-', label='A')
    ax.plot(times, Xs, 'r-', label='X')
```
